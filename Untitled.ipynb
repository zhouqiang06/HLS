```python
def sample_index(shape=(3660, 3660), rate=0.01, seed=30):
    np.random.seed(seed)
    return np.random.rand(*shape) < rate
    
```

```python
L8_bands = ['B01', 'B02', 'B03', 'B04', 'B05', 'B06', 'B07', 'B09', 'B10', 'B11', 'Fmask', 'ACmask']
S2_bands = {0: 'B01',
1: 'B02',
2: 'B03',
3: 'B04',
4: 'B05',
5: 'B06',
6: 'B07',
7: 'B08',
8: 'B8A',
9: 'B09',
10: 'B10',
11: 'B11',
12: 'B12',
13: 'ACmask',
14: 'Fmask'}

img_path_L = r'C:\Users\qzhou2\C_workdir\Goden_tile\TOA\HLS.L30.T29TQE.2021234.1.toa.hdf'
img_path_S = r'C:\Users\qzhou2\C_workdir\Goden_tile\TOA\HLS.S30.T29TQE.2021234.1.at30m.toa.hdf'

idx_map = sample_index()
samp_dict = {}
with rio.open(img_path_L) as hdf:
    # datasets = [name for name in hdf.subdatasets if name.endswith("_PAR")]
    datasets = hdf.subdatasets
    for i, dataset in enumerate(datasets):
        # Read individual layers
        with rio.open(dataset, 'r') as src:
            sample = src.read(1)[idx_map].flatten()
            layer_name = dataset.split(':')[-1]
            samp_dict[layer_name] = sample#[sample != -9999]
        print(layer_name, samp_dict[layer_name][:10])
# print(src_data.shape)
out_file = img_path_L.replace('hdf', 'sample.npz')
np.savez(out_file, **samp_dict)

samp_dict = {}
with rio.open(img_path_S) as hdf:
    # datasets = [name for name in hdf.subdatasets if name.endswith("_PAR")]
    datasets = hdf.subdatasets
    for i, dataset in enumerate(datasets):
        # Read individual layers
        with rio.open(dataset, 'r') as src:
            sample = src.read(1)[idx_map].flatten()
            layer_name = S2_bands[int(dataset.split(':')[-1])]
            samp_dict[layer_name] = sample#[sample != -9999]
        print(layer_name, samp_dict[layer_name][:10])
# print(src_data.shape)
out_file = img_path_S.replace('hdf', 'sample.npz')
np.savez(out_file, **samp_dict)
```
